```python
import youtube_dl
import pandas as pd
import os
```

```python
# playlist url
url = 'https://soundcloud.com/paul-metz-6/sets/mp'
```

```python
# create directory
savedir = "mp"
if not os.path.exists(savedir):
    os.makedirs(savedir)
```

```python
# create options for ydl
options = {
    'ignoreerrors': True,       # ignores erros
    'format': 'bestaudio/best', # choice of quality
    'extractaudio' : True,      # only keep the audio
    'audioformat' : "mp3",      # convert to mp3 
    'outtmpl': savedir + '/%(title)s -- %(uploader)s.%(ext)s',        # name the file the ID of the video
    'noplaylist' : False,}       # only download single song, not playlist
    
```

```python
with youtube_dl.YoutubeDL(options) as ydl:
    #info_dict = ydl.extract_info(url, download=False)
    #ydl.prepare_filename(info_dict)
    #print(info_dict)
    ydl.download([url])
```
